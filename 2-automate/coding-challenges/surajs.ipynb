```python
class db(object):
_db = {
        'personal_info': {
            'name': 'FirstName LastName',
            'location': 'Your Location',
            'phone': 'Your phone number',
            'email': 'email address'
        },
        'objective': "Your objective",
        "education": [
            {
                "institution": "University Name",
                "location": "Location",
                "year": 1990,
                "degree": "degree name",
                "description": "your grade description"
            },
            {
                "institution": "University Name",
                "location": "Location",
                "year": 1990,
                "degree": "degree name",
                "description": "your grade description"
            }
        ],
        "employment": [
            {
                "role": "role name",
                "year": 1990,
                "company": "company name",
                "location": "location",
                "description": [
                    "description 1",
                    "description 2",
                ]
            },
            {
                "role": "role name",
                "year": 1990,
                "company": "company name",
                "location": "location",
                "description": [
                    "description 1",
                    "description 2",
                ]
            }
        ],
        "skills": [
            "Sample skills and abilities 1",
            "Sample skills and abilities 2"
        ],
        "activities": [
            "Sample activity 1",
            "Sample activity 2"
        ],
        "hobbies": [
            "Sample hobby 1",
            "Sample hobby 2"
        ]
    }
```

```python
import os

class ResumeGenerator(object):
        
    def __init__(self):
        self._db = db._db
    
    def getData(self, header):
        return self._db[header]
    
    def getHeader(self):
        return self._db.keys()
    
    def getFileName(self):
        return "{name}-resume.md".format(name=self._db['personal_info']['name'].split(' ')[0])
    
    def writeFile(self, f, txt):
         f.write(txt)
#         print(txt)
    
    def getValuesFromDist(self,d):
        r = ""
        for k,v in d.items():
            r += v+"\r\n"
        return r
    
    def getValuesFromList(self,l, p=""):
        r = ""
        for i in l:
            r += "{prefix}{data}\r\n".format(prefix=p+" ",data=i)
        return r
    
    def create(self):
        fileName = self.getFileName()
        
        with open(fileName, "w") as w:
            self.writeFile(w,"# Resume\r\n")
            
            for k1, v1 in self._db.items():
                if(k1 == 'personal_info'):
                    self.writeFile(w,self.getValuesFromDist(self.getData(k1)))
                    self.writeFile(w,"\r\n")
                elif(k1 == 'objective'):
                    self.writeFile(w,"## {title}\r\n".format(title=k1.title()))
                    self.writeFile(w,v1+"\r\n")
                    self.writeFile(w,"\r\n")
                elif(k1 == 'education'):
                    self.writeFile(w,"## {title}\r\n".format(title=k1.title()))
                    for i in v1:
                        self.writeFile(w,"#### {institution}, {location}, {year}\r\n".format(institution=i["institution"].title(), location=i["location"].title(), year=str(i["year"])))
                        self.writeFile(w,"- {degree}\r\n".format(degree=i["degree"]))
                        self.writeFile(w,"- {degree}\r\n".format(degree=i["description"]))
                        self.writeFile(w,"\r\n")
                elif(k1 == 'employment'):
                    self.writeFile(w,"## {title}\r\n".format(title=k1.title()))
                    for i in v1:
                        self.writeFile(w,"#### {role}, {year}\r\n".format(role=i["role"].title(),year=str(i["year"])))
                        self.writeFile(w,"{company}, {location}\r\n".format(company=i["company"].title(),location=i["location"].title()))
                        self.writeFile(w, self.getValuesFromList(l=i["description"],p="- "))
                        self.writeFile(w,"\r\n")
                else:
                    self.writeFile(w,"## {title}\r\n".format(title=k1.title()))
                    self.writeFile(w, self.getValuesFromList(l=v1,p="- "))
                    self.writeFile(w,"\r\n")
```

```python
rg = ResumeGenerator()
rg.create()
```
