```python
class db(object):
    _db = {
        'personal_info': {
            'name': 'FirstName LastName',
            'location': 'Your Location',
            'phone': 'Your phone number',
            'email': 'email address'
        },
        'objective': "Your objective",
        "education": [
            {
                "institution": "University Name",
                "location": "Location",
                "year": 1990,
                "degree": "degree name",
                "description": "your grade description"
            },
            {
                "institution": "University Name",
                "location": "Location",
                "year": 1990,
                "degree": "degree name",
                "description": "your grade description"
            }
        ],
        "employment": [
            {
                "role": "role name",
                "year": 1990,
                "company": "company name",
                "location": "location",
                "description": [
                    "description 1",
                    "description 2",
                ]
            },
            {
                "role": "role name",
                "year": 1990,
                "company": "company name",
                "location": "location",
                "description": [
                    "description 1",
                    "description 2",
                ]
            }
        ],
        "skills": [
            "Sample skills and abilities 1",
            "Sample skills and abilities 2"
        ],
        "activities": [
            "Sample activity 1",
            "Sample activity 2"
        ],
        "hobbies": [
            "Sample hobby 1",
            "Sample hobby 2"
        ]
    }
```

```python
import os

class ResumeGenerator(object):
        
    def __init__(self, db):
        self._db = db
    
    def getData(self, header):
        return self._db[header]
    
    def getHeaders(self):
        return self._db.keys()
    
    def getFileName(self):
        return "{name}-resume.md".format(name=self._db['personal_info']['name'].split(' ')[0])
    
    def writeString(self, f, txt):
        r = f
        r += txt
        return r
    
    def getHeader(self, message, level=1):
        """
        Return a markdown heading
        """
        if level in range(1,6):
            return "{0} {1}\n".format("#"*level, message)
        return none
    
    def getList(self, l, t = "u"):
        """
        arguments:
            l = list
            t = type : u => unordered and o => ordered
        """
        r = ""
        if isinstance(l,(list, set, tuple)):
            if (t =="u"):
                for i in l:
                    r += "- {t}\n".format(t=i)
            elif(t=="o"):
                for i in enumerate(l):
                    r += "{o}. {t}\n".format(o=i[0],t=i[1])
            elif(t=="n"):
                for i in l:
                    r += "{t}\n".format(t=i)
        elif isinstance(l,dict):
            if (t =="u"):
                for i in l.values():
                    r += "- {t}\n".format(t=i)
            elif(t=="o"):
                for i in enumerate(l.values()):
                    r += "{o}. {t}\n".format(o=i[0],t=i[1])
            elif(t=="n"):
                for i in l.values():
                    r += "{t}\n".format(t=i)
        return r
    
    def prepareResume(self):
        w=""
        w = self.writeString(w,self.getHeader("Resume",1))            
        for k1, v1 in self._db.items():
            if(k1 == 'personal_info'):
                w = self.writeString(w,self.getList(self.getData(k1),'n'))
                w = self.writeString(w,"\n")
            elif(k1 == 'objective'):
                w = self.writeString(w,self.getHeader(k1.title(),2))
                w = self.writeString(w,v1+"\n")
                w = self.writeString(w,"\n")
            elif(k1 == 'education'):
                w = self.writeString(w,self.getHeader(k1.title(),2))
                for i in v1:
                    w = self.writeString(w,self.getHeader("{institution}, {location}, {year}".format(institution=i["institution"].title(), location=i["location"].title(), year=str(i["year"])),4))
                    w = self.writeString(w,self.getList([i["degree"],i["description"]],'u'))
                    w = self.writeString(w,"\n")
            elif(k1 == 'employment'):
                w = self.writeString(w,self.getHeader(k1.title(),2))
                for i in v1:
                    w = self.writeString(w,self.getHeader("{role}, {year}".format(role=i["role"].title(),year=str(i["year"])),4))
                    w = self.writeString(w,"{company}, {location}\n".format(company=i["company"].title(),location=i["location"].title()))
                    w = self.writeString(w, self.getList(i["description"],'u'))
                    w = self.writeString(w,"\n")
            else:
                w = self.writeString(w,self.getHeader(k1.title(),2))
                w = self.writeString(w, self.getList(v1,'u'))
                w = self.writeString(w,"\n")
        return w
    
    def writeFile(self, w, txt):
        w.write(txt)
        
    def create(self,option="w"):
        fileName = self.getFileName()
        resume = self.prepareResume()
        
        if ( option == "w"):
            with open(fileName,'w') as w:
                self.writeFile(w,resume)
        else:
            print(resume)
            
```
